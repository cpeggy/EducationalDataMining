```python
import pandas as pd
import re
import numpy as np


# 目標欄位
target_cols = [
    '自我練習平均成績','自我測驗卷數', '自我練習題目數', '完成老師指派影片數',
    '自我點播影片數', '累積影片總時數'
]

df_reduced = mathLiYu714501_df[['學校名稱'] + target_cols].copy()

def ultra_robust_convert_to_numeric(df, cols):
    for col in cols:
        # 處理所有可能異常值（包含全形、半形、特殊字元）
        def clean_number(x):
            x = str(x).replace('，', '').replace(',', '').replace('．', '.').strip()
            x = re.sub(r'[^\d\.]', '', x)
            try:
                return float(x) if x else 0.0
            except:
                return 0.0
        df[col] = df[col].apply(clean_number)
    return df


# 篩選函數
def filter_non_zero(df, cols):
    condition = (df[cols] != 0).any(axis=1)
    return df[condition]

# 超強轉型
df_numeric = ultra_robust_convert_to_numeric(df_reduced.copy(), target_cols)

# 篩選非零資料
filtered_df = filter_non_zero(df_numeric, target_cols)
```
