```python
# 篩選數學科
mathLiYu714501_df = sheetsLiYu714501[sheetsLiYu714501['科目'] == '數學']
```

```python
# 按照年級拆分並且加上 .copy()
grade7LiYu714501_math = mathLiYu714501_df[mathLiYu714501_df['年級'] == "7"].copy()
grade8LiYu714501_math = mathLiYu714501_df[mathLiYu714501_df['年級'] == "8"].copy()
grade9LiYu714501_math = mathLiYu714501_df[mathLiYu714501_df['年級'] == "9"].copy()
```

```python
# 你的目標欄位
target_cols = [
    '指派測驗平均成績', '老師指派測驗數', '老師指派測驗完成數', '老師指派測驗題目數',
    '自我練習平均成績', '自我測驗卷數', '自我練習題目數',
    '完成老師指派影片數', '自我點播影片數', '累積影片總時數'
]

# 明確資料轉換函數
def convert_to_numeric(df, cols):
    for col in cols:
        df.loc[:, col] = pd.to_numeric(df[col], errors='coerce').fillna(0)
    return df
```

```python
# 進行資料轉換
grade7LiYu714501_math = convert_to_numeric(grade7LiYu714501_math, target_cols)
grade8LiYu714501_math = convert_to_numeric(grade8LiYu714501_math, target_cols)
grade9LiYu714501_math = convert_to_numeric(grade9LiYu714501_math, target_cols)
```

```python
# 篩選非0資料函數
def filter_non_zero(df, cols):
    condition = (df[cols] != 0).any(axis=1)
    return df[condition]
```

```python
# 執行篩選
grade7LiYu714501_math_f = filter_non_zero(grade7LiYu714501_math, target_cols)
grade8LiYu714501_math_f = filter_non_zero(grade8LiYu714501_math, target_cols)
grade9LiYu714501_math_f = filter_non_zero(grade9LiYu714501_math, target_cols)
```
